```python
import matplotlib
#pyplot is a subplot 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Plot a line graph using numpy
import numpy as np
x=np.array([2,8])
y=np.array([2,80])
#plot() function is used to mark points
plt.plot(x,y)
#plt.show()
```

```python
#Plot only points
import numpy as np
x=np.array([0,5])
y=np.array([0,50])
#plot() function is used to mark points
plt.plot(x,y,'*',color='blue')
plt.show()
```

```python
#Plot multiple points
import numpy as np
x=np.array([0,2,6,10,14,18])
y=np.array([0,20,30,40,50,60])
#plot() function is used to mark points
plt.plot(x,y,'o',color='blue')
plt.show()
```

```python
import numpy as np
x=np.array([0,1,2,3,4,5])
y=np.array([0,10,20,30,40,50])
#plot() function is used to mark points
#mec- marker edge color , mfc=marker face color
plt.plot(x,y,'o',marker='*' ,ms=20, mec='g',mfc='r')
#plt.show()
```

```python
x = [3,9,12,15,18]
y = [21,30,38,49,61]
#ls- linestyle  c -  line color  lw - line width
plt.plot(x,y,ls='dotted', c='b',lw=5)
plt.show()
```

# Pie Charts

```python
data = [10,20,30,80]
mylabels= ['color1','color2','color3','color4']
plt.pie(data,labels=mylabels,startangle=50)

plt.show()
```

```python
data = [10,20,30,40]
mylabels= ['color1','color2','color3','color4']
exp_value=[0, 0, 0, 0.2]
mycolor=['purple', 'cyan','yellow','blue']
plt.pie(data,labels=mylabels,explode=exp_value,shadow = True,colors=mycolor)
plt.legend(title = 'colors')
plt.show()
```
